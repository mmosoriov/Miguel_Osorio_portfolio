#  Does the inter-industry structure of the US economy in a given year allow it to meet external final demand? If so find a production vector that will meet it exactly.


```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load Data



```python
# Load the USE matrix 
use_df = pd.read_csv('use_2017.csv',header=None)
print("USE Matrix dimensions:", use_df.shape)
print("USE Matrix head:\n", use_df.head())
```

```python
# Load the Make matrix
make_df = pd.read_csv('make_2017.csv', header=None)
print("MAKE Matrix dimensions:", make_df.shape)
print("MAKE Matrix head:\n", make_df.head())
```

```python
# Load Real Industry output table
indoutput_df = pd.read_csv('REAL_INDOUTPUT.csv', header=None)
print("REAl_INDOUTPUT TABLE dimensions:", indoutput_df.shape)
print("REAl_INDOUTPUT TABLE head:\n", indoutput_df.head())
```

```python
# Load Real commodity output table
comoutput_df = pd.read_csv('REAL_COMOUTPUT.csv', header=None)
print("REAl_COMOUTPUT TABLE dimensions:", comoutput_df.shape)
print("REAl_COMDOUTPUT TABLE head:\n", comoutput_df.head())
```

```python
# Convert DataFrames to NumPy arrays
use_matrix = use_df.values
make_matrix = make_df.values
indoutput_matrix = indoutput_df.values
comoutput_matrix = comoutput_df.values


# take headers off
use_matrix = use_matrix[1:, 1:]
make_matrix = make_matrix[1:, 1:]
indoutput_matrix = indoutput_matrix[1:, 1:]
comoutput_matrix = comoutput_matrix[1:, 1:]
```

# LEONTIEF EQUATION
## (I-DB)g = De
## We are solving for production vector gcalc( total industry output)
gcalc = L*d

gcalc = (I - (V'$\hat{q}$' ^-1)(U'$\hat{g}$'^-1))^-1 * (V'$\hat{q}$'^-1)* e


```python
# Extract V (make matrix: commodities by commodities, rows 0-165, columns 0-165)
V = make_matrix[0:166, 0:166]
print("\nV (Make Matrix):")
print("Shape:", V.shape)  # Should be (166, 166)
print("Values:\n", V)


```

```python
# Extract U (intermediate portion: commodities by industries, rows 0-165, columns 0-165)
U = use_matrix[0:166, 0:166]
print("\nU (Intermediate Use Matrix):")
print("Shape:", U.shape)  # Should be (166, 166)
print("Values:\n", U)
```

```python


# Extract the Total commodity output vector for 2017 (column 20 in zero-based indexing)
q = comoutput_matrix[0:166, 20]
print("\nq (Total commodity output vector for 2017)")
print("Shape:", q.shape)  # Should be (166, 1)
print("q_2017 (Commodity Output for 2017):", q)
```

```python
# Extract the Total industry output vector for 2017 (column 20 in zero-based indexing)
g = indoutput_matrix[0:166, 20]
print("\ng (Total industry output vector for 2017)")
print("Shape:", g.shape)  # Should be (166, 1)
print("q_2017 (industry Output for 2017):", g)
```

```python
# Extract the final demand vector for 2017 (column 167 from use matrix)
e = use_matrix[0:166, 166]
print("\ne (final demand vector for 2017)")
print("Shape:", e.shape)  # Should be (166, 1)
print("e_2017 (final demand for 2017):", e)
```

```python

# Replace zeros in g and q with a small value to avoid division by zero, cast to float
g_safe = np.where(g == 0, 1e-10, g).astype(float)
q_safe = np.where(q == 0, 1e-10, q).astype(float)

```

Market share matrix = Make matrix * inverse of total commodity output vector(square matrix, with elements in main diagonal)
D =  V '$\hat{q}$'^-1



```python


# Construct diagonal square matrice for q
q_hat = np.diag(q_safe)
# Calculate the inverse of q_hat safely
q_hat_inv = np.diag(1 / q_safe) 
# Market share matrix = Make matrix * q_hat_inv , @ for matrix multiplication
D = V.astype(float) @ q_hat_inv
```

Direct inputs coefficient matrix = Intermediates of USE matrix * inverse of total industry output vector(square matrix, with elements in main diagonal)
#
B =  U '$\hat{g}$'^-1

```python

# Construct diagonal square matrix for g
g_hat = np.diag(g_safe)
# Calculate the inverse of g_hat safely
g_hat_inv = np.diag(1 / g_safe)
# Direct inputs coefficients matrix = U * g_hat_inv
B = U.astype(float) @ g_hat_inv

```

# Consumption Matrix = Direct inputs coefficient Matrix * Market Share Matrix
# C = DB


```python

C = B @ D
```

# Leontief inverse matrix = (identity matrix - consumption matrix)^-1
L = (I-C)^-1

```python
# Create the identity matrix of size 166x166
I = np.eye(166)
# L = (I-C)^-1
L = np.linalg.inv(I - C)

```

# Total comodity output = Leontief inverse matrix * final demand
qcalc = L * e

```python

q_calc = L @ e.astype(float)

```

# Total industry output = market share matrix * total commodity output
gcalc = D * qcalc

```python


g_calc = D @ q_calc
```

```python


# Verification
print("g (original):", g)
print("g_calc:", g_calc)

#check for negative values in g_calc
if np.any(g_calc < 0):
    print("The economy is not productive because g_calc contains negative values.")
else:
    print("The economy is productive because g_calc does not contain negative values.")
```

```python

print("q (original):", q)
print("q_calc:", q_calc)
```

# Example Sector: Computing infrastructure providers, data processing, web hosting, and related services

```python



# Compute 2033 production (similar to 2017)
g_safe_2033 = np.where(g_2033 == 0, 1e-10, g_2033)
q_safe_2033 = np.where(q_2033 == 0, 1e-10, q_2033)
D_2033 = V_2033 @ np.diag(1 / q_safe_2033)
B_2033 = U_2033 @ np.diag(1 / g_safe_2033)
C_2033 = B_2033 @ D_2033
L_2033 = np.linalg.inv(np.eye(166) - C_2033)
q_calc_2033 = L_2033 @ e_2033
g_calc_2033 = D_2033 @ q_calc_2033

# Interactive visualization function taken and adapted from https://plotly.com/python/
def update_plot(sector_name):
    clear_output(wait=True)
    sector_num = [k for k, v in sector_map.items() if v.lower() == sector_name.lower()]
    if not sector_num:
        print("Sector not found. Please try again.")
        return
    sector_idx = sector_num[0] - 1  # Adjust for 0-based indexing
    
    # Data for the selected sector
    g_2017_val = g[sector_idx]
    g_2033_val = g_calc_2033[sector_idx]
    q_2017_val = q[sector_idx]
    q_2033_val = q_calc_2033[sector_idx]
    intermediate_use = U_2033[sector_idx, :].sum()
    final_demand = e_2033[sector_idx]
    
    # Create subplots
    fig = make_subplots(rows=1, cols=3, subplot_titles=("Production Comparison", "Demand Breakdown", "Output Trend"),
                        specs=[[{"type": "bar"}, {"type": "pie"}, {"type": "scatter"}]])
    
    # Bar chart: Production comparison
    fig.add_trace(go.Bar(x=["2017 Industry", "2033 Industry", "2017 Commodity", "2033 Commodity"],
                         y=[g_2017_val, g_2033_val, q_2017_val, q_2033_val], name="Production"),
                  row=1, col=1)
    
    # Pie chart: Demand breakdown
    fig.add_trace(go.Pie(labels=["Intermediate Use", "Final Demand"], values=[intermediate_use, final_demand],
                         name="Demand Breakdown"), row=1, col=2)
    
    # Line chart: Trend (using available years; extend to 2033 with prediction)
    years = [2017, 2018, 2019, 2020, 2021, 2022, 2023, 2033]
    g_trend = list(indoutput_df.values[sector_idx + 1, 20:]) + [g_2033_val]  # 2017-2023 + 2033
    fig.add_trace(go.Scatter(x=years, y=g_trend, mode="lines+markers", name="Industry Output Trend"),
                  row=1, col=3)
    
    # Update layout
    fig.update_layout(title_text=f"Production Analysis for {sector_name} (Sector {sector_num[0]})",
                      height=400, width=1200, showlegend=True)
    fig.show()

# UPDATE SECTOR NAME Computing infrastructure providers, data processing, web hosting, and related services
sector_input = widgets.Text(value="Computing infrastructure providers, data processing, web hosting, and related services", description="Sector Name:", layout=Layout(width="50%"))
update_button = widgets.Button(description="Update Plot", button_style="success")

# Define button click event
def on_button_clicked(b):
    update_plot(sector_input.value)

update_button.on_click(on_button_clicked)

# Display widgets and initial plot
display(widgets.HBox([sector_input, update_button]))
update_plot(sector_input.value)
```
